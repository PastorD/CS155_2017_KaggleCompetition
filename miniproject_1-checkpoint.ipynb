```python
%matplotlib notebook

import os
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

os.chdir("C:/Users/jmelis/Desktop/CS155 machine learning data mining/miniproject_1")

test_data = pd.read_csv('test_2008.csv')
train_data = pd.read_csv('train_2008.csv')

#x_test = test_data.iloc[0:(len(test_data)-1)]

N_cols = train_data.shape[1]
N_rows = train_data.shape[0]

x_train = np.zeros((N_rows,1))
#x_train_cat = np.zeros((1,1))
y_train = np.zeros((N_rows,1))

for i in np.arange(N_cols):
    current_col = np.transpose(np.array([train_data[train_data.columns[i]]]))
    vals, indices = np.unique(current_col, return_inverse=True)
    idx = np.transpose(np.array([indices]))
    n_cat = vals.max;
    if len(vals) > 1 and len(vals) < 101: # maximum of 10 categories per column
        if i < (N_cols-1):
            x_train = np.append(x_train, idx, 1)
            #x_train_cat np.append(x_train_cat, n_cat, 0)
        elif i == (N_cols-1):
            y_train = current_col

# 5-fold cross validation:
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, test_size=0.2, random_state=0)

'''
# random forest:
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Train random forest classifier, calibrate on validation data and evaluate
# on test data
clf1 = RandomForestClassifier(n_estimators=200,criterion="entropy")
clf1.fit(X_train, Y_train.ravel())

print("random forest")
print(accuracy_score(Y_train.ravel(), clf1.predict(X_train)))
print(accuracy_score(Y_test.ravel(), clf1.predict(X_test)))

clf1_probs_train = clf1.predict_proba(X_train)
clf1_probs_test = clf1.predict_proba(X_test)
'''

# AdaBoost 1:
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

clf1 = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1),algorithm="SAMME.R",n_estimators=200)

clf1.fit(X_train, Y_train.ravel())
print("adaboost")
print(accuracy_score(Y_train.ravel(), clf1.predict(X_train)))
print(accuracy_score(Y_test.ravel(), clf1.predict(X_test)))

clf1_probs_train = clf1.predict_proba(X_train)
clf1_probs_test = clf1.predict_proba(X_test)

# AdaBoost 2:
#from sklearn.ensemble import AdaBoostClassifier
#from sklearn.tree import DecisionTreeClassifier

clf2 = AdaBoostClassifier(DecisionTreeClassifier(max_depth=2),algorithm="SAMME.R",n_estimators=200)

clf2.fit(X_train, Y_train.ravel())
print("adaboost")
print(accuracy_score(Y_train.ravel(), clf2.predict(X_train)))
print(accuracy_score(Y_test.ravel(), clf2.predict(X_test)))

clf2_probs_train = clf2.predict_proba(X_train)
clf2_probs_test = clf2.predict_proba(X_test)

# AdaBoost 3:
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

clf3 = AdaBoostClassifier(DecisionTreeClassifier(max_depth=3),algorithm="SAMME.R",n_estimators=200)

clf3.fit(X_train, Y_train.ravel())
print("adaboost")
print(accuracy_score(Y_train.ravel(), clf3.predict(X_train)))
print(accuracy_score(Y_test.ravel(), clf3.predict(X_test)))

clf3_probs_train = clf3.predict_proba(X_train)
clf3_probs_test = clf3.predict_proba(X_test)

# Train a neural network on the probabilities of the previous 3 cycles:
from sklearn.neural_network import MLPClassifier

Y_pred_train = clf1_probs_train
Y_pred_train = np.append(Y_pred_train, clf2_probs_train, 1)
Y_pred_train = np.append(Y_pred_train, clf3_probs_train, 1)

Y_pred_test = clf1_probs_test
Y_pred_test = np.append(Y_pred_test, clf2_probs_test, 1)
Y_pred_test = np.append(Y_pred_test, clf3_probs_test, 1)

clf4 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(3, 3), random_state=1)
clf4.fit(Y_pred_train, Y_train.ravel())

print("MLP on probability scores:")
print(accuracy_score(Y_train.ravel(), clf4.predict(Y_pred_train)))
print(accuracy_score(Y_test.ravel(), clf4.predict(Y_pred_test)))

'''
# Nearest neighbor classifier:
from sklearn.neighbors import KNeighborsClassifier

clf3 = KNeighborsClassifier()
clf3.fit(X_train, Y_train.ravel())
print("Nearest Neighbors")
print(accuracy_score(Y_train.ravel(), clf3.predict(X_train)))
print(accuracy_score(Y_test.ravel(), clf3.predict(X_test)))

clf3_probs_train = clf3.predict_proba(X_train)
clf3_probs_test = clf3.predict_proba(X_test)

#combine probabilities of both classifiers and pick values with highest probability:

Y_pred_train = clf1_probs_train
Y_pred_train = np.append(Y_pred_train, clf2_probs_train, 1)
Y_pred_train = np.append(Y_pred_train, clf3_probs_train, 1)

Y_pred_test = clf1_probs_test
Y_pred_test = np.append(Y_pred_test, clf2_probs_test, 1)
Y_pred_test = np.append(Y_pred_test, clf3_probs_test, 1)

# train random forest algorithm on Y_pred:
#from sklearn.neural_network import MLPClassifier

clf4 = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
clf4.fit(Y_pred_train, Y_train.ravel())

print("MLP on probability scores:")
print(accuracy_score(Y_train.ravel(), clf4.predict(Y_pred_train)))
print(accuracy_score(Y_test.ravel(), clf4.predict(Y_pred_test)))
'''
```
